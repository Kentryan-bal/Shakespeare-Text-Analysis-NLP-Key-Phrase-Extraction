## Data Source
Shakespeare plays from the Project Gutenberg collection via NLTK's Gutenberg corpus.

## Step 1: Data Loading and Selection
Loading Shakespeare plays from the NLTK Gutenberg corpus. The corpus includes seven plays: Hamlet, Macbeth, Julius Caesar, King Lear, Romeo and Juliet, A Midsummer Night's Dream, and The Merchant of Venice.

```python
import nltk
nltk.download('gutenberg')

from nltk.corpus import gutenberg

plays = [
    "shakespeare-hamlet.txt",
    "shakespeare-macbeth.txt",
    "shakespeare-caesar.txt",
    "shakespeare-lear.txt",
    "shakespeare-romeo_and_juliet.txt",
    "shakespeare-midsummer_night.txt",
    "shakespeare-merchant_of_venice.txt"
]


print("Available Shakespeare Plays (Project Gutenberg):\n")
for i, play in enumerate(plays, start=1):
    print(f"{i}. {play.replace('shakespeare-', '').replace('.txt','').replace('_',' ').title()}")

choice = int(input("\nChoose ONE play (enter number): "))

selected_play = plays[choice - 1]

text = gutenberg.raw(selected_play)

print("\nPlay successfully loaded into variable: text")
print(text[:100])
```

### View Complete Text
Display the full text of the selected play for reference.

```python
# Run to show full printed text
print (text)
```

## Step 2: Environment Setup
Importing necessary NLP libraries and downloading required NLTK datasets for tokenization, stemming, and lemmatization.

```python
# Run this to import libraries needed
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.util import ngrams
from collections import Counter
import string

nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

## Step 3: Tokenization
Breaking down the text into individual tokens (words and punctuation). Tokenization is the foundational step for all subsequent NLP analysis, converting raw text into structured data.

```python
# Place code for Tokenization below here, do not delete this comment
tokens = word_tokenize(text)

print("Tokens:")
print(tokens)
```

## Step 4: Stemming
Applying the Porter Stemmer algorithm to reduce words to their root forms. Stemming uses heuristic rules to chop off word endings, which may result in non-dictionary words but ensures consistent grouping of related terms.

```python
# Place code for Stemming below here, do not delete this comment
stemmer = PorterStemmer()
stems = [stemmer.stem(token) for token in tokens]
print("\nStems:")
print(stems)
```

## Step 5: Lemmatization
Using WordNet Lemmatizer to reduce words to their base dictionary forms. Unlike stemming, lemmatization uses vocabulary and morphological analysis to return actual dictionary words, providing more linguistically accurate normalization.

```python
# Place code for Lemmatization below here, do not delete this comment
lemmatizer = WordNetLemmatizer()
lemmas = [lemmatizer.lemmatize(token) for token in tokens]
print("\nLemmas:")
print(lemmas)
```

## Step 6: Key Phrase Extraction Using N-Gram Analysis
Extracting candidate key phrases through statistical frequency analysis:
- **Monograms (1-grams)**: Single words that appear frequently
- **Bigrams (2-grams)**: Two-word sequences that may represent important concepts
- **Trigrams (3-grams)**: Three-word sequences that capture longer phrases

The extraction process involves:
1. Text preprocessing (lowercasing, punctuation removal)
2. Lemmatization for normalization
3. N-gram generation
4. Frequency counting to identify the most significant phrases

```python
# Place code for Key Phrase Extraction below here, do not delete this comment

# Clean tokens (lowercase, remove punctuation)
clean_tokens = [token.lower() for token in tokens if token not in string.punctuation]

# Lemmatize clean tokens
lemmas = [lemmatizer.lemmatize(token) for token in clean_tokens]

# Monograms (1-gram)
monograms = list(ngrams(lemmas, 1))

# Bigrams (2-gram)
bigrams = list(ngrams(lemmas, 2))

# Trigrams (3-gram)
trigrams = list(ngrams(lemmas, 3))

# Count frequencies
monogram_freq = Counter(monograms)
bigram_freq = Counter(bigrams)
trigram_freq = Counter(trigrams)

print("Candidate Keyphrases (Monograms):")
for phrase, freq in monogram_freq.most_common(10):
    print(" ".join(phrase), "", freq)

print("\nCandidate Keyphrases (Bigrams):")
for phrase, freq in bigram_freq.most_common(10):
    print(" ".join(phrase), "", freq)

print("\nCandidate Keyphrases (Trigrams):")
for phrase, freq in trigram_freq.most_common(10):
    print(" ".join(phrase), "", freq)
```

## Analysis Results and Insights

### 1. Vocabulary Size Analysis
Calculating the number of unique tokens to understand the lexical diversity of the text.

```python
# 1) How many unique tokens were you able to extract?
unique_token_count = len(set(tokens))
print(f"Number of unique tokens extracted: {unique_token_count}")
```

### 2. Top Key Phrases by Category
Identifying the most frequent phrases across all n-gram categories to understand the text's dominant themes and linguistic patterns.

**Monograms (Single Words):**

```python
# 2) What are the top 10 Key Phrases in your assigned play

print("Top 10 Candidate Keyphrases (Monograms):")
for phrase, freq in monogram_freq.most_common(10):
    print(" ".join(phrase), "", freq)
```

**Bigrams (Two-Word Phrases):**

```python
print("\nTop 10 Candidate Keyphrases (Bigrams):")
for phrase, freq in bigram_freq.most_common(10):
    print(" ".join(phrase), "", freq)
```

**Trigrams (Three-Word Phrases):**

```python
print("\nTop 10 Candidate Keyphrases (Trigrams):")
for phrase, freq in trigram_freq.most_common(10):
    print(" ".join(phrase), "", freq)
```

### 3. Contextual Analysis of Key Phrases
Locating high-frequency key phrases within the original text to verify their significance and understand their usage context. This helps validate that our extracted phrases are truly meaningful in the context of the play.

```python
# 3) Choose one key phrase from your output and locate it in the original text.
import re

# Choose a meaningful keyphrase (e.g., top bigram: "my lord")
chosen_phrase = "my lord"

print(f"Searching for key phrase: '{chosen_phrase}'\n")

pattern = re.compile(re.escape(chosen_phrase), re.IGNORECASE)
matches = list(pattern.finditer(text))

print(f"Found {len(matches)} occurrences in the original text.\n")

for i, match in enumerate(matches[:3], 1):
    start = max(0, match.start() - 50)
    end = min(len(text), match.end() + 50)
    context = text[start:end].replace('\n', ' ')
    print(f"Occurrence {i}: ...{context}...")
```

### 4. N-Gram Distribution Summary
Analyzing the distribution of different n-gram types in the top key phrases to understand which phrase lengths are most significant in the text.

```python
# 4) How many monograms, bigrams, trigrams, and n-grams appeared in the top 10 keyphrases

# Count from each category's top 10
mono_top10 = monogram_freq.most_common(10)
bi_top10 = bigram_freq.most_common(10)
tri_top10 = trigram_freq.most_common(10)

print("Top 10 from each n-gram category:")
print(f"  - Monograms (1-word): {len(mono_top10)}")
print(f"  - Bigrams (2-word): {len(bi_top10)}")
print(f"  - Trigrams (3-word): {len(tri_top10)}")
print(f"\nTotal keyphrases analyzed: {len(mono_top10) + len(bi_top10) + len(tri_top10)}")
```
